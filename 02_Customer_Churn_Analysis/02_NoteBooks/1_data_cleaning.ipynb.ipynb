```python
# ========================
# DATA CLEANING
#=========================
# 1. Convert date column
# 2. Check duplicates
# 3. Handle missing values
# 4. Fix Yes/No column
# 5. Check datatypes
# 6. Save cleaned dataset
```

```python
df["signup_date"] = pd.to_datetime(df["signup_date"])
df.info()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["annual_income"] = df["annual_income"].fillna(df["annual_income"].median())

df["customer_satisfaction"] = df["customer_satisfaction"].fillna(
    df["customer_satisfaction"].median()
)

df["num_complaints"] = df["num_complaints"].fillna(
    df["num_complaints"].median()
)

df["avg_monthly_gb"] = df["avg_monthly_gb"].fillna(
    df["avg_monthly_gb"].median()
)

df["credit_score"] = df["credit_score"].fillna(
    df["credit_score"].median()
)
```

```python
df.isnull().sum()
```

```python
df["paperless_billing"].unique()
```

```python
df["paperless_billing"] = df["paperless_billing"].map({
    "Yes": 1,
    "No": 0
})
```

```python
df.to_csv(
    r"C:\Users\boyal\OneDrive\Customer_Churn_Analysis\Data\cleaned_data.csv",
    index=False
)
```
