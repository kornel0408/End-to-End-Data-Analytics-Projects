```python
# ==================================================================
#          ADVANCED EXPLORATORY DATA ANALYSIS (EDA)
# =================================================================
# 1. Correlation heatmap
# 2. Churn vs Gender
# 3. Churn vs Payment method
# 4. Churn vs Internet service
# 5. Churn vs Number of services
# 6. Churn vs Late payments
# 7. Churn vs Service calls
# 8. Churn vs Avg monthly GB
# 9. Pair insights / strong features
```

```python
# Correlation heatmap to see relationship between numeric columns

plt.figure(figsize=(12,8))

sns.heatmap(
    df.corr(numeric_only=True),   # correlation of numeric columns
    cmap="coolwarm"               # color style
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
# Check churn based on gender

sns.countplot(
    x="gender",
    hue="churn",
    data=df
)

plt.title("Churn vs Gender")
plt.show()
```

```python
# Check churn based on payment method

plt.figure(figsize=(8,5))

sns.countplot(
    x="payment_method",
    hue="churn",
    data=df
)

plt.xticks(rotation=30)

plt.title("Churn vs Payment Method")

plt.show()
```

```python
# Check churn based on internet service

sns.countplot(
    x="has_internet_service",
    hue="churn",
    data=df
)

plt.title("Churn vs Internet Service")

plt.show()
```

```python
# Check churn based on number of services used

sns.boxplot(
    x="churn",
    y="num_services",
    data=df
)

plt.title("Churn vs Number of Services")

plt.show()
```

```python
# Check churn based on late payments

sns.boxplot(
    x="churn",
    y="late_payments",
    data=df
)

plt.title("Churn vs Late Payments")

plt.show()
```

```python
# Check churn based on number of service calls

sns.boxplot(
    x="churn",
    y="num_service_calls",
    data=df
)

plt.title("Churn vs Service Calls")

plt.show()
```

```python
# Check churn based on data usage

sns.boxplot(
    x="churn",
    y="avg_monthly_gb",
    data=df
)

plt.title("Churn vs Avg Monthly GB")

plt.show()
```
